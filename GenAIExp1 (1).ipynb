```python
import openai as OpenAI

client = OpenAI.OpenAI(api_key="YOUR_API_KEY")
```

```python
response = client.chat.completions.create(model= "gpt-4o-mini",
                                            messages=[{"role" : "user",
                                                       "content" : "Explain c# data types in detail."
                                                       }
                                                      ]
                                          )

print(response.choices[0].message.content)
```
