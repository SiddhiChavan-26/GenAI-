```python
from groq import Groq

client = Groq(api_key="YOUR_API_KEY")
```

```python
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "Hello, what is AI?"}
    ]
)

print(response.choices[0].message.content)
```

```python
import re

def calculator(expression):
    try:
        # Remove words like "calculate", "solve", etc.
        cleaned = re.sub(r"[a-zA-Z=]", "", expression)

        result = eval(cleaned)
        return f"The result is: {result}"

    except:
        return "Invalid expression"
```

```python
def intelligent_agent(user_query):

    keywords = ["+", "-", "*", "/", "%"]

    if any(op in user_query for op in keywords):
        return calculator(user_query)

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "user", "content": user_query}
        ]
    )

    return response.choices[0].message.content
```

```python
print(intelligent_agent("calculate 10 + 20 * 3"))
```

```python
print("🤖 Intelligent Agent Chatbot Started (type 'exit' to stop)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Agent: Goodbye!")
        break

    response = intelligent_agent(user_input)
    print("Agent:", response)
    print()
```
